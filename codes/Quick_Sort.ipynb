```python
import time, sys, numpy as np, matplotlib.pyplot as plt

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr)//2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

sizes = [10, 100, 500, 1000, 2000]
times, spaces = [], []

for n in sizes:
    arr = np.random.randint(0, 10000, n).tolist()
    start = time.time()
    quick_sort(arr)
    end = time.time()
    times.append(end-start)
    spaces.append(sys.getsizeof(arr))

plt.figure(figsize=(8,5))
plt.plot(sizes, times, 'o-', label='Quick Sort')
plt.xlabel('Input Size (n)')
plt.ylabel('Time (seconds)')
plt.title('Quick Sort: Time Complexity')
plt.grid(True)
plt.show()

plt.figure(figsize=(8,5))
plt.plot(sizes, spaces, 'o-', color='red')
plt.xlabel('Input Size (n)')
plt.ylabel('Estimated Space (bytes)')
plt.title('Quick Sort: Space Complexity')
plt.grid(True)
plt.show()
```
