```python
import time, sys, numpy as np, matplotlib.pyplot as plt

def binary_search(arr, target):
    low, high = 0, len(arr)-1
    while low <= high:
        mid = (low + high)//2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

sizes = [10, 100, 1000, 5000, 10000]
times, spaces = [], []

for n in sizes:
    arr = sorted(np.random.randint(0, 10000, n).tolist())
    start = time.time()
    binary_search(arr, arr[n//2])
    end = time.time()
    times.append(end-start)
    spaces.append(sys.getsizeof(arr))

plt.figure(figsize=(8,5))
plt.plot(sizes, times, 'o-', color='navy')
plt.xlabel('Input Size (n)')
plt.ylabel('Time (seconds)')
plt.title('Binary Search: Time Complexity')
plt.grid(True)
plt.show()

plt.figure(figsize=(8,5))
plt.plot(sizes, spaces, 'o-', color='crimson')
plt.xlabel('Input Size (n)')
plt.ylabel('Estimated Space (bytes)')
plt.title('Binary Search: Space Complexity')
plt.grid(True)
plt.show()
```
