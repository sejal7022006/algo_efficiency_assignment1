```python
import time, sys, numpy as np, matplotlib.pyplot as plt

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

sizes = [10, 100, 500, 1000]
times, spaces = [], []

for n in sizes:
    arr = np.random.randint(0, 10000, n).tolist()
    start = time.time()
    bubble_sort(arr)
    end = time.time()
    times.append(end-start)
    spaces.append(sys.getsizeof(arr))

plt.figure(figsize=(8,5))
plt.plot(sizes, times, 'o-', color='brown')
plt.xlabel('Input Size (n)')
plt.ylabel('Time (seconds)')
plt.title('Bubble Sort: Time Complexity')
plt.grid(True)
plt.show()

plt.figure(figsize=(8,5))
plt.plot(sizes, spaces, 'o-', color='orange')
plt.xlabel('Input Size (n)')
plt.ylabel('Estimated Space (bytes)')
plt.title('Bubble Sort: Space Complexity')
plt.grid(True)
plt.show()
```
