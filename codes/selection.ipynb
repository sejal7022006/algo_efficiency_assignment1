```python
import time, sys, numpy as np, matplotlib.pyplot as plt

def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

sizes = [10, 100, 500, 1000]
times, spaces = [], []

for n in sizes:
    arr = np.random.randint(0, 10000, n).tolist()
    start = time.time()
    selection_sort(arr)
    end = time.time()
    times.append(end-start)
    spaces.append(sys.getsizeof(arr))

plt.figure(figsize=(8,5))
plt.plot(sizes, times, 'o-', color='teal')
plt.xlabel('Input Size (n)')
plt.ylabel('Time (seconds)')
plt.title('Selection Sort: Time Complexity')
plt.grid(True)
plt.show()

plt.figure(figsize=(8,5))
plt.plot(sizes, spaces, 'o-', color='magenta')
plt.xlabel('Input Size (n)')
plt.ylabel('Estimated Space (bytes)')
plt.title('Selection Sort: Space Complexity')
plt.grid(True)
plt.show()
```
