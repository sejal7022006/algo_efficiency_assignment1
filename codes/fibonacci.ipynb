```python
import time, sys
import numpy as np
import matplotlib.pyplot as plt

def fib_recursive(n):
    if n <= 1:
        return n
    return fib_recursive(n-1) + fib_recursive(n-2)

def fib_dp(n):
    if n <= 1:
        return n
    dp = [0]*(n+1)
    dp[1] = 1
    for i in range(2, n+1):
        dp[i] = dp[i-1] + dp[i-2]
    return dp[n]

def measure(func, n):
    start = time.time()
    func(n)
    end = time.time()
    return end - start, sys.getsizeof(n)

sizes = [5, 10, 20, 25, 30]
rec_times, dp_times, rec_space, dp_space = [], [], [], []

for n in sizes:
    t, s = measure(fib_recursive, n)
    rec_times.append(t); rec_space.append(s)
    t, s = measure(fib_dp, n)
    dp_times.append(t); dp_space.append(s)

plt.figure(figsize=(8,5))
plt.plot(sizes, rec_times, 'o-', label='Recursive')
plt.plot(sizes, dp_times, 'o-', label='Dynamic Programming')
plt.xlabel('n'); plt.ylabel('Time (seconds)')
plt.title('Fibonacci: Time Complexity')
plt.legend(); plt.grid(True)
plt.show()

plt.figure(figsize=(8,5))
plt.plot(sizes, rec_space, 'o-', label='Recursive')
plt.plot(sizes, dp_space, 'o-', label='Dynamic Programming')
plt.xlabel('n'); plt.ylabel('Estimated Space (bytes)')
plt.title('Fibonacci: Space Complexity')
plt.legend(); plt.grid(True)
plt.show()
```
