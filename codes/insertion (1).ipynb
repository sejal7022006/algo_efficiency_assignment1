```python
import time, sys, numpy as np, matplotlib.pyplot as plt

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j >= 0 and key < arr[j]:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key

sizes = [10, 100, 500, 1000]
times, spaces = [], []

for n in sizes:
    arr = np.random.randint(0, 10000, n).tolist()
    start = time.time()
    insertion_sort(arr)
    end = time.time()
    times.append(end-start)
    spaces.append(sys.getsizeof(arr))

plt.figure(figsize=(8,5))
plt.plot(sizes, times, 'o-', color='green')
plt.xlabel('Input Size (n)')
plt.ylabel('Time (seconds)')
plt.title('Insertion Sort: Time Complexity')
plt.grid(True)
plt.show()

plt.figure(figsize=(8,5))
plt.plot(sizes, spaces, 'o-', color='purple')
plt.xlabel('Input Size (n)')
plt.ylabel('Estimated Space (bytes)')
plt.title('Insertion Sort: Space Complexity')
plt.grid(True)
plt.show()
```
